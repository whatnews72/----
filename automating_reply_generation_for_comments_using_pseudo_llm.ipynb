```python
import random

def pseudo_llm(input_text):
    # 0또는 1 중에서 랜덤한 값을 반환
    output = random.choice(['더 노력하겠습니다.', '의견 감사합니다.'])
    return output

```

```python
import urllib.request
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix

def generate_reply(input_text):
    output = pseudo_llm(input_text)
    return output

urllib.request.urlretrieve("https://raw.githubusercontent.com/tykimos/tykimos.github.io/master/warehouse/dataset/tarrr_sample_submit.txt",
                           filename="tarrr_sample_submit.txt")
# 파일을 DataFrame으로 로드
df = pd.read_csv('tarrr_sample_submit.txt', delimiter='\t')

for index, row in df.iterrows():
    comment = row['comment']
    reply = generate_reply(comment)

    print(f"[{index}]")
    print("comment : ", comment)
    print("reply : ", reply)
    print("----------------")
```
