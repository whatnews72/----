# Convolution Layer

```python
import tensorflow as tf
import pandas as pd
```

```python
# 1. 과거의 데이터를 준비합니다.                                            # 필터셋은 몇 개을 사용할 것인가
(독립, 종속), _ = tf.keras.datasets.mnist.load_data()                       # 필터셋의 사이즈를 얼마로 할 것인가
print(독립.shape, 종속.shape)
독립 = 독립.reshape(60000, 28, 28, 1)  #컨볼루션은 데이터가 3차원이어야 하기 때문에 reshape해준
종속 = pd.get_dummies(종속)        #원핫 인코딩 해준다(?)
print(독립.shape, 종속.shape)
```

```python
# 2. 모델의 구조를 만듭니다.
X = tf.keras.layers.Input(shape=[28, 28, 1])
H = tf.keras.layers.Conv2D(3, kernel_size=5, activation='swish')(X)   #필터 3장
H = tf.keras.layers.Conv2D(6, kernel_size=5, activation='swish')(H)   #필터 6장
H = tf.keras.layers.Flatten()(H)                #데이터를 한 줄로 펴서 표형태로 만들어줌
H = tf.keras.layers.Dense(84, activation='swish')(H)
Y = tf.keras.layers.Dense(10, activation='softmax')(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# 모델을 학습하고
model.fit(독립, 종속, epochs=10)
```

```python
# 모델을 이용합니다.
pred = model.predict(독립[0:5])
pd.DataFrame(pred).round(2)
```
