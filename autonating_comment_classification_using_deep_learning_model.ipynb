```python
from transformers import pipeline

pipe = pipeline("text-classification", model="sangrimlee/bert-base-multilingual-cased-nsmc", device = 0)

def dl_model(input_text):
  output = pipe(input_text)
  if output[0]['label'] == 'positive':
    return 1
  else:
    return 0
```

```python
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/tykimos/tykimos.github.io/master/warehouse/dataset/tarr_train.txt", filename="tarr_train.txt")
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix

def classify_text(input_text):
  output = dl_model(input_text)
  return output

  # ÌååÏùºÏùÑ DataFrameÏúºÎ°ú Î°úÎìú
df = pd.read_csv('tarr_train.txt', delimiter='\t')

actual_labels = []
predicted_labels = []

for index, row in df.iterrows():
  comment = row['comment']
  actual_label = row['label']
  predicted_label = classify_text(comment)
  actual_labels.append(actual_label)
  predicted_labels.append(predicted_label)

  print("Comment:", comment)
  print("actual class:", actual_label)
  print("Predicted class:", predicted_label)
  print("------------------------")

# Ï†ïÌôïÎèÑ Í≥ÑÏÇ∞
accuracy = accuracy_score(actual_labels, predicted_labels)
print(f"Accuracy: {accuracy*100:.2f}%")

print(set(actual_labels))
print(set(predicted_labels))

# ÌòºÎèô ÌñâÎ†¨ Í≥ÑÏÇ∞
cm = confusion_matrix(actual_labels, predicted_labels, labels=[1,0])

# ÌòºÎèô ÌñâÎ†¨ Ï∂úÎ†•
print("\nConfusion Matrix:")
print("         Predicted:")
print("         Í∏çÏ†ï    Î∂ÄÏ†ï")
print("Actual")
print("Í∏çÏ†ï    {:<5}  {:<5}".format(cm[0][0], cm[0][1]))
print("Î∂ÄÏ†ï    {:<5}  {:<5}".format(cm[1][0], cm[1][1]))
```
